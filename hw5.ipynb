There is a very low standard deviation, showing that there isn't much fluctuation between folds. This means the model performs consistently across different pieces of data and thus good generalization. In first glance, the mean doesn't seem too great, however the dummyclassifier only scored 0, so while the linear model doesn't have the best score, it is a really good improvement from the DummyClassifier but there is still a lot of room for improvement.

````python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.pipeline import make_pipeline

lr_pipeline = make_pipeline(preprocessor, LogisticRegression(max_iter=1000))

param_grid = {
    'logisticregression__C': [0.01, 0.1, 1, 10, 100]
}

gs = GridSearchCV(lr_pipeline, 
                  param_grid = param_grid, #Make grid search object
                  scoring = 'f1',
                  n_jobs=-1, 
                  return_train_score=True
                 )
gs.fit(X_train, y_train) #Fit grid search object

lr_best = gs.best_estimator_ 
lr_predict = lr_best.predict(X_test)

precision_lr = precision_score(y_test, lr_predict)
recall_lr = recall_score(y_test, lr_predict) #Get best recall and prediction scores on the Logistic regression model that has the best parameters

print ("Best Recall Score: ", recall_lr)
print ("Best Precision Score: ", precision_lr)

print("Best hyperparameters:", gs.best_params_)
````

````python
cv_score_f1 = cross_val_score(lr_pipeline, X_train, y_train, cv=5, scoring='f1')
print ("SD F1: ", np.std(cv_score_f1)) #Instead of scoring on accuracy, I am scoring on f1
print ("Mean F1: ", np.mean(cv_score_f1))
````
